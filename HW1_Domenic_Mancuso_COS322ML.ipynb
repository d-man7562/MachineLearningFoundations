### **Basic Pandas commands**
Please learn and practice the following pandas commands by running the code, changing inputs, and observing the output.

df.head(n): view the first n rows

df.tail(m): view the last m rows

df.shape: check the shape of the dataset (rows, columns)

df.columns: see column names

df.info(): check data types and non-null counts

df.describe(): summary statistics for numeric columns

df['species'].value_counts(): value counts for a categorical column

df.iloc[row_index, column_index]: select rows and columns by integer position, and the index always starts at 0

df.loc[row_label, column_label]: select rows and columns using their names, not their numerical positions; filter by condition


````python
# Examples that demonstrate how to use .iloc[] (index-based selection) and .loc[] (label-based selection)
# to access rows and columns in a DataFrame


# .iloc[], indexing Starts from 0

# Get the first row
iris.iloc[0]
````

````python
# .iloc[n:m] returns a DataFrame of rows from position n to m-1.
# This follows Python-style slicing, where the end index is exclusive.
# Example: df.iloc[2:5] returns rows at positions 2, 3, and 4.

# Get the first 5 ro
iris.iloc[:5]
````

````python
# Get the first column
iris.iloc[:, 0]
````

````python
# Get first 3 rows and first 2 columns
df.iloc[0:3, 0:2]
````

````python
# Get all rows where species == 'setosa'
iris.loc[iris['species'] == 'setosa']
````

````python
# Get all rows where petal_length > 5 and only select 'petal_length' and 'species'
iris.loc[iris['petal_length'] > 5, ['petal_length', 'species']]
````
